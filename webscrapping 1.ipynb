```python
#Write a python program to scrape mentioned details from dineout.co.in 
!pip install bs4
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
soup
```

```python
#scrapping name

restaurant_name = []
for i in soup.find_all('div', class_="restnt-info cursor"):
    restaurant_name.append(i.text)
```

```python
restaurant_name
```

```python
#location
location = []
for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)
```

```python
location
```

```python
#cuisine & rates
cuisine = []
for i in soup.find_all('span', class_="double-line-ellipsis"):
    cuisine.append(i.text)
```

```python
cuisine
```

```python
#image URL
images = []
for i in soup.find_all("img", class_="no-img"):
    images.append(i['data-src'])
images
```

```python
print(len(restaurant_name), len(location),len(cuisine), len(images))
```

```python
#adding the data into the data frame

df= pd.DataFrame({'restaurant_name' :restaurant_name, 'location':location,'cuisine':cuisine,'images':images})
```

```python
df
```
